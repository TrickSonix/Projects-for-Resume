```python
def fix_rooms_zeros(df, mean_sqaure_by_rooms):
    df.loc[df['Rooms'] == 0, 'Rooms'] = np.ceil(df['Square']*1.37/mean_sqaure_by_rooms.loc[1, 'Square'])
    return df
```

```python
def fix_small_square(df, mean_sqaure_by_rooms):
    for index, value in df.loc[df['Square'] < df['Rooms']*15, 'Rooms'].iteritems():
        try:
            df.loc[index, 'Square'] = mean_sqaure_by_rooms.loc[mean_sqaure_by_rooms['Rooms'] == value, 'Square'].iloc[0]
        except IndexError:
            df.loc[index, 'Square'] = mean_sqaure_by_rooms['Square'].mean()
    return df
```

```python
X_train = fix_small_square(X_train, mean_sqaure_by_rooms)
```

```python
X_train.loc[X_train['Square'] > X_train['Rooms']*100]
```

```python
def fix_big_square(df, mean_sqaure_by_rooms):
    for index, value in df.loc[df['Square'] > df['Rooms']*100, 'Rooms'].iteritems():
        try:
            df.loc[index, 'Square'] = mean_sqaure_by_rooms.loc[mean_sqaure_by_rooms['Rooms'] == value, 'Square'].iloc[0]
        except IndexError:
            df.loc[index, 'Square'] = mean_sqaure_by_rooms['Square'].mean()
    return df
```

```python
X_train = fix_big_square(X_train, mean_sqaure_by_rooms)
```

```python
def fillna_life_square(df, mean_life_sqaure_by_rooms):
    for index, value in df.loc[df['LifeSquare'].isna(), 'Rooms'].iteritems():
        try:
            df.loc[index, 'LifeSquare'] = mean_life_sqaure_by_rooms.loc[mean_life_sqaure_by_rooms['Rooms'] == value, 'LifeSquare'].iloc[0]
        except IndexError:
            df.loc[index, 'LifeSquare'] = mean_life_sqaure_by_rooms['LifeSquare'].mean()
    return df
```

```python
def fix_big_life_square(df, mean_life_sqaure_by_rooms):
    for index, value in df.loc[df['LifeSquare'] > df['Rooms']*100, 'Rooms'].iteritems():
        try:
            df.loc[index, 'LifeSquare'] = mean_life_sqaure_by_rooms.loc[mean_life_sqaure_by_rooms['Rooms'] == value, 'LifeSquare'].iloc[0]
        except IndexError:
            df.loc[index, 'LifeSquare'] = mean_life_sqaure_by_rooms['LifeSquare'].mean()
    return df
```

```python
X_train = fix_big_life_square(X_train, mean_life_sqaure_by_rooms)
```

```python
def fix_small_life_square(df, mean_life_sqaure_by_rooms):
    for index, value in df.loc[df['LifeSquare'] < df['Rooms']*10, 'Rooms'].iteritems():
        try:
            df.loc[index, 'LifeSquare'] = mean_life_sqaure_by_rooms.loc[mean_life_sqaure_by_rooms['Rooms'] == value, 'LifeSquare'].iloc[0]
        except IndexError:
            df.loc[index, 'LifeSquare'] = mean_sqaure_by_rooms['LifeSquare'].mean()
    return df
```

```python
stats = X_train.groupby(['DistrictId', 'Rooms'], as_index=False)[['Square']].mean().rename(columns={'Square':'mean_square'})
stats.head()
```

```python
def join_stats_mean_square(df, stats, df_source, mode='train'):
    df = pd.merge(df, stats, on=['DistrictId', 'Rooms'], how='left')
    if mode == 'test':
        df['mean_square'] = df['mean_square'].fillna(df_source['mean_square'].mean())
    return df
```

```python
X_train = join_stats_mean_square(X_train, stats, X_train)
```

```python
X_train.info()
```

```python
stats2 = X_train.groupby(['DistrictId', 'HouseYear'], as_index=False)[['Square']].mean().rename(columns={'Square':'mean_square_hy'})
stats2.head()
```

```python
def join_stats_mean_square_hy(df, stats2, df_source, mode='train'):
    df = pd.merge(df, stats2, on=['DistrictId', 'HouseYear'], how='left')
    if mode == 'test':
        df['mean_square_hy'] = df['mean_square_hy'].fillna(df_source['mean_square_hy'].mean())
    return df
```

```python
X_train = join_stats_mean_square_hy(X_train, stats2, X_train)
```

```python
def data_prepare(df, mean_life_sqaure_by_rooms, mean_sqaure_by_rooms, stats, stats2, df_source, mode='train'):
    df = fix_wrong_rooms_small(df)
    df = fix_small_square(df, mean_sqaure_by_rooms)
    df = fix_big_square(df, mean_sqaure_by_rooms)
    df = fillna_life_square(df, mean_life_sqaure_by_rooms)
    df = fix_big_life_square(df, mean_life_sqaure_by_rooms)
    df = fix_small_life_square(df, mean_life_sqaure_by_rooms)
    df = join_stats_mean_square(df, stats, df_source, mode=mode)
    df = join_stats_mean_square_hy(df, stats2, df_source, mode=mode)
    return df
```
